# Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('breast+cancer+wisconsin+diagnostic\wdbc.data')
```

```python
data
```

```python
cols = ['ID', 'Diagnosis', 'radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1', 'compactness1', 'concavity1', 'concave points1',
         'symmetry1', 'fractal_dimension1','radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2', 'compactness2', 'concavity2',
         'concave_points2', 'symmetry2', 'fractal_dimension2', 'radius3', 'texture3', 'perimeter3', 'area3', 'smoothness3', 'compactness3',
         'concavity3', 'concave_points3', 'symmetry3', 'fractal_dimension3'
        ]
len(cols)
```

```python
data.columns = cols
```

```python
data.head()
```

```python
data.shape
```

```python
# Converting Diagnosis columns to binary
data['Diagnosis'] = (data['Diagnosis'] == 'M').astype(int)
```

```python
data.head()
```

```python
#Plotting radius1 with respect to Diagnosis to visualize their relationship
colors = data['Diagnosis'].map({1: 'orange', 0: 'Blue'})

plt.figure(figsize=(14, 6))
plt.scatter(data.index, data['radius1'], c=colors)
plt.xlabel('Index')
plt.ylabel('radius1')
plt.legend = True
```

```python
#Plotting every column and to weed out redundant ones
for i in data.columns[2:]:
    plt.figure(figsize=(14, 6))
    plt.scatter(data.index, data[i], c=colors, label=i,)
    plt.xlabel('Index')
    plt.ylabel(i)
    plt.show()
```

### Reducing Noise

```python
corr_target = data.corr()["Diagnosis"].drop(labels=['Diagnosis', 'ID'])
corr_target.sort_values()
```

Dropping columns with less than 0.1 correlation

```python
noise_cols = ['fractal_dimension1', 'texture2', 'smoothness2', 'symmetry2', 'fractal_dimension2']

data.drop(noise_cols, axis=1, inplace=True)
```

```python
data.shape
```

```python
#Checking for bias in dataset
data["Diagnosis"].value_counts()
```

### Fixing Bias towards benign by resampling

```python
from sklearn.utils import resample, shuffle
```

```python
M_data = data[data['Diagnosis'] == 1]
B_data = data[data['Diagnosis'] == 0]

B_resampled = resample(B_data, replace=False, n_samples=len(M_data), random_state=42)

data_resampled = pd.concat([B_resampled, M_data])
data_resampled = shuffle(data_resampled).reset_index(drop=True)
```
